```python
time.sleep(3)
# def draw a traingle

def draw_triangle(start_x,start_y,side_length):
    pg.mouseDown(start_x,start_y)
    pg.moveTo(start_x + side_length, start_y, 3)
    pg.mouseUp()
    pg.mouseDown(start_x + side_length, start_y)
    pg.moveTo(start_x + side_length / 2, start_y + side_length * (3**0.5) / 2, 3)
    pg.mouseUp()
    pg.mouseDown(start_x + side_length / 2, start_y + side_length * (3**0.5) / 2)
    pg.moveTo(start_x, start_y, 3)
    pg.mouseUp()

draw_triangle(300,300,200)
```

```python


def draw_line(x, y, x1, y1, step_size=10):
    pg.moveTo(x, y)
    time.sleep(3)

    dx = x1 - x
    dy = y1 - y

    mag = (dx**2 + dy**2)**0.5

    dx /= mag
    dy /= mag

    steps = int(mag / step_size)

    pg.mouseDown()

    for _ in range(steps):
        pg.moveRel(dx * step_size, dy * step_size, duration=0)

    pg.mouseUp()


draw_line(300, 300, 600, 300)
```
